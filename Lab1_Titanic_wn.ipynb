## EXPLORACIÓN DE DATATRAIN

```python
datatrain.info()
```

```python
# Selección de variables
train = datatrain.drop(['Ticket', 'Cabin', 'Name'], axis = 1)
train.head()
```

```python
train.info()
```

```python
train.shape
```

```python
#Revisión de nulos
train.isnull().sum()
```

#Ingenieria de Caracteristicas TRAIN

```python
#One hot encoding de SEX
ohe = pd.get_dummies(train['Sex'], drop_first=True)
train.drop('Sex', axis=1, inplace=True)
train = train.join(ohe)
train.head()
```

```python
#One hot encoding de EMBARKED
ohe = pd.get_dummies(train['Embarked'], drop_first=True)
train.drop('Embarked', axis=1, inplace=True)
train = train.join(ohe)
train.head()
```

```python
#Imputación de media en variable AGE
meanValue = np.round(train['Age'].mean(), 0)
```

```python
train['Age'] = train['Age'].fillna(meanValue)
train.head()
```

```python
train.isnull().values.any()
```

```python
train.isnull().sum()
```

```python
X_train = train.iloc[:, train.columns != 'Survived']
y_train= train.loc[:, 'Survived']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
std_scaler = StandardScaler()
X = std_scaler.fit_transform(X_train)
```

#KERAS

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
inputs = X_train.shape[1]
```

## Ejercicio con 2 capas ocultas y 6 neuronas

```python
X_train.shape[1]
```

```python
# Base secuencial para definir la estructura de la Red (SIEMPRE INICIAR CON SEQUENTIAL)
clasificador = Sequential()

#primera capa oculta
clasificador.add(Dense(input_dim = inputs, units=6, activation='relu', kernel_initializer='uniform'))

#segunda capa oculta
clasificador.add(Dense(units=6, activation='relu', kernel_initializer='uniform'))

#capa de salida
clasificador.add(Dense(units=1, activation='sigmoid', kernel_initializer='uniform'))
```

```python
#parametros de optimización
clasificador.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

#parametros de entrenamiento.
clasificador.fit(X_train, y_train, batch_size=25, epochs=100)
```
